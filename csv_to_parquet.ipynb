```python
import pandas as pd
from concurrent.futures import ProcessPoolExecutor
from clean_csv import clean_csv
```

```python
TRIP_FARE_PATH = "/Users/mac/Development/nyc-taxi-spark-project/trip_fare"
TRIP_DATA_PATH = "/Users/mac/Development/nyc-taxi-spark-project/trip_data"
```

```python
with ProcessPoolExecutor() as executor:
    futures = []
    for i in range(1, 13):
        input_path = f"{TRIP_FARE_PATH}/trip_fare_{i}.csv"
        output_path = f"{TRIP_FARE_PATH}_cleaned/trip_fare_{i}.csv"
        futures.append(executor.submit(clean_csv, input_path, output_path))
    
    for future in futures:
        future.result()
```

```python
with ProcessPoolExecutor() as executor:
    futures = []
    for i in range(1, 13):
        input_path = f"{TRIP_DATA_PATH}/trip_data_{i}.csv"
        output_path = f"{TRIP_DATA_PATH}_cleaned/trip_data_{i}.csv"
        futures.append(executor.submit(clean_csv, input_path, output_path))
    
    for future in futures:
        future.result()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.functions import col
import seaborn as sns
import matplotlib.pyplot as plt
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType, DoubleType, TimestampType


# Create SparkSession
spark = SparkSession.builder \
    .appName("nycTaxi") \
    .getOrCreate()

```

```python
fare_schema = StructType([
    StructField("medallion", StringType(), True),
    StructField("hack_license", StringType(), True),
    StructField("vendor_id", StringType(), True),
    StructField("pickup_datetime", TimestampType(), True),
    StructField("payment_type", StringType(), True),
    StructField("fare_amount", DoubleType(), True),
    StructField("surcharge", DoubleType(), True),
    StructField("mta_tax", DoubleType(), True),
    StructField("tip_amount", DoubleType(), True),
    StructField("tolls_amount", DoubleType(), True),
    StructField("total_amount", DoubleType(), True)
])

trip_schema = StructType([
    StructField("medallion", StringType(), True),
    StructField("hack_license", StringType(), True),
    StructField("vendor_id", StringType(), True),
    StructField("rate_code", IntegerType(), True),
    StructField("store_and_fwd_flag", StringType(), True),
    StructField("pickup_datetime", TimestampType(), True),
    StructField("dropoff_datetime", TimestampType(), True),
    StructField("passenger_count", IntegerType(), True),
    StructField("trip_time_in_secs", IntegerType(), True),
    StructField("trip_distance", DoubleType(), True),
    StructField("pickup_longitude", DoubleType(), True),
    StructField("pickup_latitude", DoubleType(), True),
    StructField("dropoff_longitude", DoubleType(), True),
    StructField("dropoff_latitude", DoubleType(), True)
])

df_trip = spark.read.option("header", True).schema(trip_schema).csv("trip_data_cleaned")
df_fare = spark.read.option("header", True).schema(fare_schema).csv("trip_fare_cleaned")
```

```python
df_fare.write.mode("overwrite").parquet("trip_fare.parquet")
```

```python
df_trip.write.mode("overwrite").parquet("trip_data.parquet")
```
